## How Python Differs from Go and TypeScript

In Go you have `go run`, `go build`, and no REPL. In TS you have `ts-node`, compiled JS, and browser console. Python has 3 modes: `.py` scripts, ipython REPL, and Jupyter notebooks. Each serves a different purpose.

## Running .py Files

```python
# This is equivalent to `go run main.go` or `ts-node script.ts`
# .py files are your production code — run from the terminal
print("hello from a .py file")
```

**Go/TS comparison:** `.py` files are your production code. Like Go, you run them from the terminal. Unlike Go, no compilation step — Python interprets directly.

```python
# Experiment: Check which Python you're running — this matters with conda
import sys
print(f"Python version: {sys.version}")
print(f"Executable: {sys.executable}")
```

## ipython REPL

```python
# ipython-style exploration
# dir() and help() are like GoDoc but live and interactive
x = [1, 2, 3]
print(f"Type: {type(x)}")
print(f"\nUseful methods on list (filtered):")
print([m for m in dir(x) if not m.startswith('_')])
```

**Go/TS comparison:** Go has no REPL. TS has `ts-node` but it's clunky. ipython is where you experiment — tab completion, `?` for help, `%timeit` for benchmarks. It's the scratch pad you'll use constantly.

```python
# Experiment: %timeit is a "magic command" — ipython/Jupyter only, not regular Python
# It runs the expression many times and reports the average
%timeit sum(range(1000))
```

## Jupyter Notebooks

**Go/TS comparison:** No equivalent in Go or TS. Notebooks mix code, output, and documentation in one file. They run on a **kernel** (a live Python process). Cells share state — a variable defined in cell 1 is available in cell 10.

```python
# Shared state demo: this variable persists across cells
name = "Kyle"
```

```python
# This works because the kernel remembers `name` from the previous cell
print(f"Hello, {name}")
```

```python
# Experiment: shared state can be a gotcha
# Try running this cell, then the cell above, then this cell again
# The value of `items` changes each time!
items = []
items.append("run")
print(f"items has {len(items)} element(s): {items}")
```

**Go/TS comparison — the out-of-order gotcha:** In Go, code runs top to bottom, always. In a notebook, you can run cell 5 before cell 3. This is both powerful and dangerous. When confused, use **Kernel → Restart & Run All** to reset to a clean, top-to-bottom state.

## Conda Environments

**Go/TS comparison:** Like `go mod` or `node_modules` — isolates dependencies per project. But conda also manages the **Python version itself** (npm only manages packages, not Node). Think of conda as `nvm` + `npm` combined.

```python
# Check which conda environment is active
import sys
print(f"Environment: {sys.prefix}")
print(f"This tells you which conda env is active")
```

```python
# Experiment: the ! prefix runs shell commands from inside a notebook
# This is like backticks in Go's os/exec or child_process in Node
!python --version
```
